```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://www.programmableweb.com/category/all/apis"
response = requests.get(url)
response
```

```python
data = response.text
soup = BeautifulSoup(data)
```

```python
table = soup.find('table')
table_body = table.find('tbody')
```

```python
table_rows = table_body.find_all('tr')
data_list = {}
row_no = 0
for tr in table_rows:
    td = tr.find_all('td')
    # col = [i.text for i in td]
    # data_list.append(col)
    
    api_name = td[0].text
    api_url = td[0].find('a').get('href')
    api_description = td[1].text
    api_category = td[2].text
    # print(api_name, api_url, api_description, api_category)
    row_no += 1
    data_list[row_no] = [api_name, api_url, api_description, api_category]
    
df = pd.DataFrame.from_dict(data_list, orient='index', columns=['API', 'URL', 'DESCRIPTION', 'CATEGORY'])
df
```

```python
url = "https://www.programmableweb.com/category/all/apis"
row_no = 0
data_list = {}

while True:
    response = requests.get(url)
    data = response.text
    soup = BeautifulSoup(data)
    table = soup.find('table')
    table_body = table.find('tbody')
    table_rows = table_body.find_all('tr')

    for tr in table_rows:
        td = tr.find_all('td')
        api_name = td[0].text
        api_url = "https://www.programmableweb.com" + td[0].find('a').get('href')
        api_description = td[1].text
        api_category = td[2].text
    # print(api_name, api_url, api_description, api_category)
        row_no += 1
        data_list[row_no] = [api_name, api_url, api_description, api_category]
    
    next_page = soup.find('a', {'title':'Go to next page'})
    if next_page.get('href'):
        url = "https://www.programmableweb.com" + next_page.get('href')
        print(url)
    else:
        break

print("Total rows : ", row_no )
    

# df = pd.DataFrame.from_dict(data_list, orient='index', columns=['API', 'URL', 'DESCRIPTION', 'CATEGORY'])
```
